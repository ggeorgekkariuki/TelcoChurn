## Data Cleaning

##### ***Churn***

```python
df.rename(columns={'Churn': 'churn_label'}, inplace=True)
df.columns
```

```python
# Indices
try:
    assert 'Churn' not in df.columns
except:
    print("The column exists already")
else:
    # Create col with churn values
    df['Churn'] = df['churn_label'].apply(lambda value: 0 if value == 'No' else 1)
```

```python
df.Churn.unique()
```

##### *Total Charges*

```python
# Records that are blank in the TotalCharges column
df[df['TotalCharges'].apply(len) < 2]['TotalCharges']

# Indices
try:
    empty_records = df[df['TotalCharges'].apply(len) < 2].index
    assert empty_records is not np.nan
except:
    print("These indices were already dropped")
else:
    # Drop these indices
    df = df.drop(index=empty_records)

    # Convert column to numerical value
    df['TotalCharges'] = pd.to_numeric(df['TotalCharges'])
```

```python
# Save the file
df.to_csv('telco_clean.csv')
```

### Feature Engineering

```python
feature_data = df.copy()
```

##### ***Tenure Grouping***

```python
df['tenure_group'] = pd.cut(df['tenure'], bins=[0, 12, 24, 48, 72], labels=['0–12', '13–24', '25–48', '49–72'])
df['tenure_group'].value_counts(normalize=True)
```

##### ***Binary Encoding***

```python
def binary_encode(label): 
    if label == 'Yes': 
        return 1 
    elif label == 'No': 
        return 0 
    else: 
        return label
```

```python
feature_data[list(binary_columns)] = feature_data[list(binary_columns)].map(binary_encode)
```

```python
# Check
for col in list(binary_columns):
    print(feature_data[col].unique())
```

##### ***Mutli Encoding***

```python

dummy_multi_data = pd.get_dummies(data= feature_data[list(multi_columns)], dtype=int, drop_first=True)
feature_data.drop(columns=list(multi_columns), inplace=True)
feature_data = pd.concat([feature_data, dummy_multi_data], axis=1)
```

### Bivariate Analysis

```python
def bivariate(col1, target='TotalCharges', rotation=False, **kwargs):
    col_totals = df.groupby(col1)[target].median()

    plt.bar(x=col_totals.index, height=col_totals.values, **kwargs)
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.title(f'Median {target} across {col1.title()}')
    plt.ylabel(f'Median {target.title()}')
    plt.xticks(rotation=45 if rotation is True else 0)
    plt.xlabel(col1.title());

def churn_rate_plot(col1):
    churn_rates = df.groupby(col1)['Churn'].mean()
    plt.bar(churn_rates.index, churn_rates.values, color='salmon')
    plt.title(f'Churn Rate by {col1.title()}')
    plt.ylabel('Churn Rate')
    plt.xticks(rotation=45)
    plt.grid(axis='y', linestyle='--', alpha=0.7)
```
